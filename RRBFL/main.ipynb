```python
import subprocess
import threading
import datetime

def read_output(name, proc):
    for line in iter(proc.stdout.readline, ''):
        print(f"{datetime.datetime.now()}-[{name}][-]: {line}", end='')  # 实时打印输出
        
def create_thread(name, command):
    # 使用 subprocess.Popen 启动进程，同时捕获标准输出和标准错误
    process = subprocess.Popen(
        command,
        stdout=subprocess.PIPE,  # 重定向标准输出到管道
        stderr=subprocess.STDOUT, # 将标准错误也重定向到标准输出
        universal_newlines=True  # 确保输出文本模式
    )
    
    # 创建一个线程来读取输出，避免阻塞主线程
    output_thread = threading.Thread(target=read_output, args=(name, process,))
    output_thread.daemon = True  # 设置为守护线程，当主线程结束时自动退出
    output_thread.start()
    return output_thread
# 定义一个函数来持续读取输出


# 创建生产者服务器进程
python_path = "../venv/Scripts/python.exe"
threads = []
t = create_thread("Miner", [python_path, "miner.py", "-g", "1", "-l", "2"])
threads.append(t)
for i in range(2):
    t = create_thread(f"Client_{i}", [python_path, "client.py", "-d", f"data/federated_data_{i}.d"])
    threads.append(t)

for t in threads:
    t.join()
```
